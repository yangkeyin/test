```python
import numpy as np
from scipy.optimize import  leastsq
import matplotlib.pyplot as plt
```

```python
array = np.random.randint(1,10, size=(2,10))
print(array)
```

```python
def real_func(x):
    return np.sin(2*np.pi*x)

def fit_func(params, x):
    return np.polyval(params, x)

def loss(params, x, y):
    return fit_func(params, x) - y
```

```python
def fitting (M=0):
    x = np.linspace(0,1,10)
    y = [np.random.normal(0,0.1) + i for i in real_func(x)]
    p_init = np.random.rand(M+1)
    func = leastsq(loss, p_init, args = (x, y))

    x_vals = np.linspace(0,1,1000)
    y_fit_vals = fit_func(func[0], x_vals)
    y_real_vals = real_func(x_vals)
    
    
    plt.plot(x_vals, y_fit_vals, label="fitting curve")
    plt.plot(x_vals, y_real_vals, label="real curve")
    plt.plot(x, y, "bo", label="noise")
    plt.title("funtion")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()

    plt.show()
```

```python
func0 = fitting(M=0)
```

```python
func1 = fitting(M=1)
```
